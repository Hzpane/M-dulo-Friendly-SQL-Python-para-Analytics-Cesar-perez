## Distribución de pedidos por ciudad

Este gráfico permite identificar las ciudades con mayor volumen de pedidos registrados dentro del dataset.

```python
# Conteo de pedidos por ciudad

city_orders = df["City"].value_counts().reset_index()

city_orders.columns = ["City", "Total_Orders"]


fig = px.bar(
    city_orders,
    x="City",
    y="Total_Orders",
    title="Cantidad de pedidos por ciudad",
    text="Total_Orders"
)

fig.update_traces(textposition="outside")

fig.show()
```

### Interpretación

Se observa la distribución de pedidos según ciudad, permitiendo identificar los mercados con mayor demanda dentro de la plataforma de delivery.

```python
payment = df["Payment_Method"].value_counts().reset_index()

payment.columns = ["Payment_Method", "Total"]


fig = px.pie(
    payment,
    names="Payment_Method",
    values="Total",
    title="Distribución de métodos de pago",
    hole=0.4
)

fig.show()
```

### Interpretación

El gráfico muestra la proporción de utilización de cada método de pago, permitiendo identificar cuáles son los medios preferidos por los usuarios.

```python
fig = px.histogram(
    df,
    x="Delivery_Duration_Minutes",
    nbins=30,
    title="Distribución del tiempo de entrega"
)

fig.show()
```

### Interpretación

La distribución permite observar el comportamiento habitual de los tiempos de entrega e identificar posibles concentraciones o valores extremos.

```python
fig = px.box(
    df,
    y="Delivery_Duration_Minutes",
    title="Distribución y valores atípicos del tiempo de entrega"
)

fig.show()
```

### Interpretación

El boxplot permite identificar la dispersión de los tiempos de entrega y detectar posibles valores extremos que podrían representar incidencias operativas.

```python
fig = px.scatter(
    df.sample(5000),
    x="Delivery_Distance_km",
    y="Delivery_Duration_Minutes",
    title="Relación entre distancia y duración de entrega",
    opacity=0.5
)

fig.show()
```

### Interpretación

Se analiza la relación entre distancia recorrida y duración del pedido. Una tendencia positiva indicaría que los pedidos con mayor distancia requieren más tiempo de entrega.

```python
import duckdb

# Registramos el DataFrame como una tabla SQL
duckdb.sql("CREATE TABLE orders AS SELECT * FROM df")
```

```python
# Mostrar las 10 primeras filas
query1 = duckdb.query("""
SELECT Order_ID, User_ID, City, Total_Price, Delivery_Duration_Minutes
FROM orders
LIMIT 10
""").to_df()

query1
```

```python
query2 = duckdb.query("""
SELECT Order_ID, City, Delivery_Duration_Minutes, Total_Price
FROM orders
WHERE City = 'Dubai' AND Delivery_Duration_Minutes > 60
LIMIT 20
""").to_df()

query2
```

```python
query3 = duckdb.query("""
SELECT City, COUNT(*) AS Total_Orders, AVG(Delivery_Duration_Minutes) AS Avg_Duration
FROM orders
GROUP BY City
ORDER BY Total_Orders DESC
""").to_df()

query3
```

```python
query4 = duckdb.query("""
SELECT Order_ID, User_ID, Total_Price, City
FROM orders
ORDER BY Total_Price DESC
LIMIT 10
""").to_df()

query4
```

```python
query5 = duckdb.query("""
SELECT City, Payment_Method, AVG(Total_Price) AS Avg_Price, COUNT(*) AS Total_Orders
FROM orders
WHERE Total_Price > 50
GROUP BY City, Payment_Method
ORDER BY Avg_Price DESC
""").to_df()

query5
```
